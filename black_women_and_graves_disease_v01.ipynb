```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the Aesthetic Archetype
try:
    plt.style.use('seaborn-v0_8-whitegrid')
except:
    plt.style.use('seaborn-whitegrid')

custom_palette = ["#8B0000", "#2F4F4F", "#4682B4", "#A9A9A9", "#800000"]
sns.set_palette(custom_palette)

print("Aesthetic Archetype: MADMall Deep Narrative Loaded.")
```

## Data Synthesis & Initial Insights

```python
np.random.seed(42)
n_records = 2000

races = ['White', 'Black', 'Asian', 'Hispanic', 'Other']
race_distribution = [0.5, 0.3, 0.1, 0.05, 0.05]
data = {
    'PatientID': range(1, n_records + 1),
    'Age_at_Diagnosis': np.random.normal(loc=40, scale=14, size=n_records).astype(int).clip(18, 85),
    'Race': np.random.choice(races, size=n_records, p=race_distribution),
    'Gender': np.random.choice(['Female', 'Male'], size=n_records, p=[0.85, 0.15]),
    'Income_Level': np.random.choice(['Low', 'Medium', 'High'], size=n_records, p=[0.4, 0.4, 0.2])
}
df = pd.DataFrame(data)

# Disparity logic simulation
df['Graves_Diagnosis'] = 0
fem_idx = df[df['Gender'] == 'Female'].index
df.loc[fem_idx, 'Graves_Diagnosis'] = np.random.choice([0, 1], size=len(fem_idx), p=[0.4, 0.6])
black_fem_idx = df[(df['Gender'] == 'Female') & (df['Race'] == 'Black')].index
df.loc[black_fem_idx, 'Graves_Diagnosis'] = np.random.choice([0, 1], size=len(black_fem_idx), p=[0.2, 0.8])

diag_idx = df[df['Graves_Diagnosis'] == 1].index
df['Outcome'] = 'No Diagnosis'
for idx in diag_idx:
    race = df.loc[idx, 'Race']
    rem_p = 0.65 if race == 'White' else 0.4
    if df.loc[idx, 'Income_Level'] == 'Low': rem_p -= 0.15
    df.loc[idx, 'Outcome'] = np.random.choice(['Remission', 'Persistent'], p=[max(0.1, rem_p), 1-max(0.1, rem_p)])

# Create numerical success metric for plotting
df['Remission_Success'] = (df['Outcome'] == 'Remission').astype(int)
```

```python
# VISUALIZING THE CHASM
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

diagnosed_df = df[df['Graves_Diagnosis'] == 1].copy()

sns.kdeplot(data=diagnosed_df, x='Age_at_Diagnosis', hue='Race', fill=True, palette=custom_palette, ax=ax[0])
ax[0].set_title('THE AGE OF DIAGNOSIS: A Racial Overlay', fontsize=16, fontweight='bold', color='#2F4F4F')

sns.pointplot(data=diagnosed_df, x='Income_Level', y='Remission_Success', hue='Race',
              palette=custom_palette, ax=ax[1], order=['Low', 'Medium', 'High'])
ax[1].set_title('THE REMISSION CHASM: Income vs. Race', fontsize=16, fontweight='bold', color='#8B0000')
ax[1].set_ylabel('Success Probability')
ax[1].axhline(0.5, color='#8B0000', alpha=0.3, linestyle='--')

sns.despine()
plt.show()
```

# Act IV: The Brains of MADMall - A Custom Model Family
## Modeling & Fine-Tuning: Beyond Generic AI

The centerpiece of the MADMall architecture is a **Family of Custom Models**. These models represent the **Single Source of Truth** for the entire platform.

---

```python
from sklearn.ensemble import RandomForestClassifier
m_df = df[df['Graves_Diagnosis'] == 1].copy()
m_df = pd.get_dummies(m_df, columns=['Race', 'Income_Level'])
X = m_df.drop(columns=['PatientID', 'Outcome', 'Graves_Diagnosis', 'Gender', 'Remission_Success'])
y = m_df['Remission_Success']

clf = RandomForestClassifier(n_estimators=500, random_state=42).fit(X, y)
print(f"MADMall Model Family Success Probability: {clf.score(X, y):.2f}")
```
